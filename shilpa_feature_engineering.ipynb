```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
home_data = pd.read_csv('C:\\Users\\shgupta\\OneDrive - Five9\\Desktop\\analytica\\train.csv')
home_data.head()

# Create a FacetGrid plot
g = sns.FacetGrid(home_data, col='OverallQual', row='YearBuilt')
g = g.map(sns.scatterplot, 'GrLivArea', 'SalePrice')

# Feature Engineering
# 1. Age of the house
home_data['Age'] = home_data['YrSold'] - home_data['YearBuilt']

# 2. Total Square Footage of all floors
home_data['TotalSF'] = home_data['TotalBsmtSF'] + home_data['1stFlrSF'] + home_data['2ndFlrSF']

# 3. Total number of rooms
room_cols = ['FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd']
home_data['TotalRooms'] = home_data[room_cols].sum(axis=1)

# 4. Year of last renovation
home_data['YearRemodAdd'] = home_data[['YearBuilt', 'YearRemodAdd']].max(axis=1)

# 5. Lot size
home_data['Log Lot Area'] = np.log(home_data['Lot Area'])

plt.show()
```
