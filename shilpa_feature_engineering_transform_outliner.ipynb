```python
import numpy as np
import pandas as pd

# Load the data from the csv file
home_data = pd.read_csv("C:\\Users\\shgupta\\OneDrive - Five9\\Documents\\Analytica\\train.csv")

# Log transform the SalePrice column
home_data["SalePrice"] = np.log(home_data["SalePrice"])

# View the histogram of the transformed SalePrice column
home_data["SalePrice"].hist()
```

```python
import seaborn as sns
from scipy.stats import norm
from scipy import stats
import matplotlib.pyplot as plt


sns.distplot(home_data['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(home_data['SalePrice'], plot=plt)
```

```python
import seaborn as sns
from scipy import stats

# Plot the distribution of the SalePrice variable
sns.distplot(home_data['SalePrice'], fit=stats.norm);

# Get the fitted parameters for the normal distribution
(mu, sigma) = stats.norm.fit(home_data['SalePrice'])

# Print the fitted parameters
print('mu = {:.2f} and sigma = {:.2f}'.format(mu, sigma))
```
