```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Load data
home_data = pd.read_csv('C:\\Users\\shgupta\\OneDrive - Five9\\Desktop\\analytica\\train.csv')

# Check number of samples in data
n_samples = len(home_data)
if n_samples == 0:
    raise ValueError("Dataset is empty")

# Split data into training and testing sets
test_size = 0.2
train_data, test_data = train_test_split(home_data, test_size=test_size)

# Check number of samples in training set
n_train_samples = len(train_data)
if n_train_samples == 0:
    raise ValueError("Training set is empty")

# Check number of samples in testing set
n_test_samples = len(test_data)
if n_test_samples == 0:
    raise ValueError("Testing set is empty")
```
